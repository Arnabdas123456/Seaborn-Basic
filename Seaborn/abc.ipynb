HeatMap plot

```python
from matplotlib import pyplot as plt 
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
arr2=np.linspace(1,10,20).reshape(4,5)
print(arr2)
```

```python
y={'fontsize':15,'color':'r'}
v=sns.heatmap(arr2,vmin=0,vmax=20,cmap="PuOr",annot=True,annot_kws=y,linewidth=4,linecolor='y')
v.set(xlabel="python",ylabel="java")
```

```python
data=sns.load_dataset('anagrams')
x=data.drop('attnr',axis=1).head(10)
print(x)
```

```python
sns.heatmap(x,annot=True)
```

```python
arr=np.linspace(1,10,10).reshape(2,5)
arr
```

```python
sns.heatmap(arr,vmin=0,vmax=10,cmap="gist_heat",annot=True)
```

```python
z=np.array([['a0','a1','a2','a3','a4'],
            ['b0','b1','b2','b3','b4']])
z
```

```python
sns.heatmap(arr,vmin=0,vmax=10,cmap="Dark2",annot=z,fmt="s")
plt.show()
```
